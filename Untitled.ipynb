```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_colwidth', 600)
```

```python
# Get the html from the page
url = "https://books.toscrape.com/index.html"
response = requests.get(url)
print(response.status_code)
response.ok
```

```python
# Making soup
soup = BeautifulSoup(response.content)
```

```python
# Defining group of all book divs
total_divs = soup.find_all("li", attrs={'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
len(total_divs)
```

```python
# Slicing book Sharp Objects
book_test = total_divs[3]
book_test
```

```python
# empty list for results (Will end up a list of dictionaries)
total_books_data = []

for book_div in total_divs:
        
    # Finding the h2 that has title class
    title = book_div.find('h3').text

     # Find the a tag with the Learn string
    title_url = book_div.find("a").attrs['href']
    
    # Get any type of tag that has class=location)
    in_stock = book_div.find(attrs={'class':'instock availability'}).text.strip()
    
    # Get the date (text from time tag)
    price = book_div.find(attrs={'class':'price_color'}).text
    
    # Putting it all together
    book_info = {'Title':title, 
                "URL":title_url, 
                "Availability":in_stock,
                "Price": price
                }
    total_books_data.append(book_info)
```

```python
# Convert list of dictionaries to a dataframe
book_df = pd.DataFrame(total_books_data)
book_df.head()
```

```python
# Our URL only has the paths within the site
# Making base URL 
base_url = 'https://books.toscrape.com/'
book_df['Complete_Link'] = base_url + book_df['URL']
book_df
```

```python
# Open link to extract description
sites = book_df['Complete_Link']
site = sites[0]
```

```python
# Get the job page's response
import time
response = requests.get(site)
time.sleep(2)
```
